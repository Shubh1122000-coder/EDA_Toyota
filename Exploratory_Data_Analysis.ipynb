```python
import pandas as pd
import numpy as np
```

```python
#Pandas data frame
car_data = pd.read_csv('toyota.csv', index_col=0,na_values=["??","???"])
```

```python
#copy the data
car_data2=car_data.copy()
```

```python
#univariate analysis, gives the count of Fuel type
pd.crosstab(index=car_data2["FuelType"],columns="count",dropna=True)
```

```python
#to find frequency count of variables automatic
pd.crosstab(index=car_data2["Automatic"],columns="count",dropna=True)
```

```python
#Bivariate Analysis
pd.crosstab(index=car_data2["Automatic"],columns=car_data2["FuelType"],dropna=True)
```

```python
#Otherway to check bivariate Analysis
pd.crosstab(index=car_data2["FuelType"],columns=car_data2["Automatic"],dropna=True)
```

```python
#Joint Probability, before doing this execute on index as fueltype
pd.crosstab(index=car_data2["FuelType"],columns=car_data2["Automatic"],dropna=True,margins=True)
```

```python
#to find marginal probability
pd.crosstab(index=car_data2["Automatic"],columns=car_data2["FuelType"],dropna=True,margins=True)
```

```python
#probability of variable automatic, same as marginal probability of row
pd.crosstab(index=car_data2["Automatic"],columns="count",dropna=True,normalize=True)
```

```python
#Conditional Probability, the probability of P(FuelType|Automatic)
pd.crosstab(index=car_data2["Automatic"],columns=car_data2["FuelType"],dropna=True,normalize=True)
```

```python
#Finding the probability of P(Automatic|FuelType)
pd.crosstab(index=car_data2["Automatic"],columns=car_data2["FuelType"],dropna=True,margins=True)
```
